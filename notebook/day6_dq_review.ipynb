```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("database/nifty100.db")

tables = pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table'
""", conn)

audit = []

for table in tables["name"]:
    df = pd.read_sql(f"SELECT * FROM {table}", conn)

    audit.append([
        table,
        len(df),
        len(df.columns),
        df.isnull().sum().sum(),
        df.duplicated().sum()
    ])

audit_df = pd.DataFrame(
    audit,
    columns=[
        "table",
        "rows",
        "columns",
        "nulls",
        "duplicates"
    ]
)

audit_df
```
