```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("database/nifty100.db")

companies = pd.read_sql(
    "SELECT * FROM companies",
    conn
)

# Remove first row containing column names
companies = companies.iloc[1:].reset_index(drop=True)

# Rename columns using first data row
companies.columns = [
    'id','company_logo','company_name','chart_link',
    'about_company','website','nse_profile',
    'bse_profile','face_value','book_value',
    'roce_percentage','roe_percentage'
]

companies.to_sql(
    "companies",
    conn,
    if_exists="replace",
    index=False
)

print("New Count:", len(companies))

conn.close()
```
