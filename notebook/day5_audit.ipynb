```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("database/nifty100.db")
```

```python
tables = pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table'
""", conn)

tables
```

```python
for table in tables["name"]:

    df = pd.read_sql(
        f"SELECT * FROM {table}",
        conn
    )

    print("\n===================")
    print(table)

    print("Rows:", len(df))
    print("Columns:", len(df.columns))

    print("Null Values:")
    print(df.isnull().sum().sum())

    print("Duplicate Rows:")
    print(df.duplicated().sum())
```

```python
audit = []

for table in tables["name"]:

    df = pd.read_sql(
        f"SELECT * FROM {table}",
        conn
    )

    audit.append({
        "table_name": table,
        "rows": len(df),
        "columns": len(df.columns),
        "null_values": df.isnull().sum().sum(),
        "duplicates": df.duplicated().sum()
    })

audit_df = pd.DataFrame(audit)

audit_df.to_csv(
    "reports/load_audit.csv",
    index=False
)

audit_df
```
